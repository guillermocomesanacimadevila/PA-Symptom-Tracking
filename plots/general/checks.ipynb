```python
keep_cols = ["model", "TARGET_MAE", "TARGET_R2", "Y_MAE", "Y_R2"]
df = df[[col for col in keep_cols if col in df.columns]]
```

```python
lasso = df[df["model"] == "lasso"]
elasticnet = df[df["model"] == "elasticnet"]
rf = df[df["model"] == "rf"]
knn = df[df["model"] == "knn"]
svr = df[df["model"] == "svr"]
xgb = df[df["model"] == "xgb"]
catboost = df[df["model"] == "catboost"]
mlp = df[df["model"] == "mlp"]
```

```python
LASSO = lasso["Y_MAE"].mean()
ELASTIC = elasticnet["Y_MAE"].mean()
RF = rf["Y_MAE"].mean()
KNN = knn["Y_MAE"].mean()
SVR = svr["Y_MAE"].mean()
XGBOOST = xgb["Y_MAE"].mean()
CATBOOST = catboost["Y_MAE"].mean()
MLP = mlp["Y_MAE"].mean()
```

```python
model_frames = {
    "LASSO": lasso,
    "ELASTICNET": elasticnet,
    "RF": rf,
    "KNN": knn,
    "SVR": svr,
    "XGBOOST": xgb,
    "CATBOOST": catboost,
    "MLP": mlp
}
```

```python
results = {name: df["Y_MAE"].mean() for name, df in model_frames.items()}
```

```python
ranked = sorted(results.items(), key=lambda x: x[1], reverse=True)
```

```python
for model, score in ranked:
    print(f"{model}: {score:.4f}")
```
